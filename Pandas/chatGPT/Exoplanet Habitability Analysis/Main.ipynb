```python
df['isHabitable'] = (
    (df['flux_received'] >= 0.3) & (df['flux_received'] <= 1.5) &
    (df['planet_radius'] >= 0.5) & (df['planet_radius'] <= 2)
)
```

```python
plt.scatter(df['star_temperature'], df['flux_received'])
plt.xlabel('Star Temperature')
plt.ylabel('Flux Received')
```

```python
plt.scatter( df['isHabitable'],df['discovery_year'])
```

```python
def assign_star_type(temp):
    if temp > 30000:
        return "O"
    elif 10000 <= temp <= 30000:
        return "B"
    elif 7500 <= temp < 10000:
        return "A"
    elif 6000 <= temp < 7500:
        return "F"
    elif 5200 <= temp < 6000:
        return "G"
    elif 3700 <= temp < 5200:
        return "K"
    elif temp < 3700:
        return "M"
    else:
        return "Unknown"

df['star_type'] = df['star_temperature'].apply(assign_star_type)
```
