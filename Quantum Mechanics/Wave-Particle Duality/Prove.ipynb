```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
screenWidth = 200
slitDistance  = 20
waveLength = 10
numParticles = 10000
```

```python
screen = np.zeros(screenWidth)
```

```python
x = np.linspace(-screenWidth / 2, screenWidth / 2, screenWidth)
```

```python
slit1Wave = np.sin(2  * np.pi * x / waveLength )
slit2Wave = np.sin(2 * np.pi * (x - slitDistance) / waveLength)
interferencePattern = (slit1Wave + slit2Wave) ** 2
```

```python
# Normalize the interference pattern to make it a probability distribution
interferencePattern = interferencePattern / np.sum(interferencePattern)
```

```python
particlePosition = np.random.choice(range(screenWidth), size=numParticles, p=interferencePattern)
```

```python
for pos in particlePosition:
    screen[pos] += 1
```

```python
screen
```

```python
plt.figure(figsize=(10, 6))
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, interferencePattern, label="Interference Pattern")
plt.title("Wave-like Behavior: Interference Pattern")
plt.xlabel("Screen Position")
plt.ylabel("Intensity")
plt.legend()
```

```python
plt.subplot(2, 1, 2)
plt.bar(x, screen, width=1, color='blue', label="Particle Hits")
plt.title("Particle-like Behavior: Detection on Screen")
plt.xlabel("Screen Position")
plt.ylabel("Particle Count")
plt.legend()
plt.tight_layout()
plt.show()
```
